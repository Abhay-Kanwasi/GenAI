```python
from langchain_openai import OpenAIEmbeddings
from langchain_chroma import Chroma
from langchain_core.documents import Document
from dotenv import load_dotenv

load_dotenv()
```

```python
doc1 = Document(
    page_content="OpenAI provides advanced language models such as GPT-4, widely used for natural language processing tasks including chatbots, summarization, and creative writing.",
    metadata={"platform": "OpenAI", "focus": "Language Models", "founded": "2015"}
)

doc2 = Document(
    page_content="Google DeepMind is known for breakthroughs in reinforcement learning and AI research, including AlphaGo and AlphaFold, which solved complex problems in games and biology.",
    metadata={"platform": "DeepMind", "focus": "Reinforcement Learning & Research", "founded": "2010"}
)

doc3 = Document(
    page_content="Anthropic develops AI systems with a strong emphasis on safety and alignment, offering models like Claude that prioritize responsible AI interactions.",
    metadata={"platform": "Anthropic", "focus": "AI Safety & Alignment", "founded": "2021"}
)

doc4 = Document(
    page_content="Microsoft Azure AI provides cloud-based AI services, including machine learning, computer vision, and natural language APIs, enabling enterprises to integrate AI into applications.",
    metadata={"platform": "Microsoft Azure AI", "focus": "Cloud AI Services", "founded": "2010"}
)

doc5 = Document(
    page_content="IBM Watson offers enterprise AI solutions for industries such as healthcare and finance, with tools for natural language understanding, data analysis, and decision support.",
    metadata={"platform": "IBM Watson", "focus": "Enterprise AI Solutions", "founded": "2011"}
)

doc6 = Document(
    page_content="Amazon Web Services (AWS) AI provides scalable machine learning services like SageMaker, as well as APIs for vision, speech, and language, supporting developers and businesses.",
    metadata={"platform": "AWS AI", "focus": "Cloud Machine Learning & APIs", "founded": "2006"}
)

docs = [doc1, doc2, doc3, doc4, doc5, doc6]
embeddings = OpenAIEmbeddings()

vector_store = Chroma(
    embedding_function=embeddings,
    persist_directory='chroma_db',
    collection_name='sample'
)
```

```python
vector_store.add_documents(docs)
```

```python
# view documents
vector_store.get(include=['embeddings', 'documents', 'metadata'])
```

```python
# search in document
vector_store.similarity_search(
    query='Who among them are oldest company?',
    k=2 #Number of output you want
)
```

```python
# search with similarity score
vector_store.similarity_search_with_score(
    query='Who among them are oldest company?',
    k=2
)
```

```python
# meta-data filtering
vector_store.similarity_search_with_score(
    query="",
    filter={"platform": "AWS AI", "focus": "Cloud Machine Learning & APIs", "founded": "2006"}
)
```

```python
# update document
update_doc1 = Document(
    page_content="Amazon Web Services (AWS) AI delivers a broad suite of artificial intelligence and machine learning tools. Key offerings include Amazon SageMaker for building and deploying ML models, and pre-trained APIs for vision, speech, and language tasks. These services enable developers and enterprises to scale AI solutions efficiently in the cloud.",
    metadata={"platform": "AWS AI", "focus": "Cloud AI & Machine Learning Services", "founded": "2006"}
)
vector_store.update_document(document_id='<your doc1 id>', document=update_doc1)
```
